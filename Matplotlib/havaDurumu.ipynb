```python
import matplotlib.pyplot as plt

edirne_sıcaklık = [24.5, 24.3, 24.3, 14.3, 9.0, 4.4] 
sinop_sıcaklık = [22.8, 23.2, 20.0, 16.2, 12.6, 9.2]
antalya_sıcaklık = [28.4, 28.3, 25.1, 20.5, 15.4, 11.6]
kars_sıcaklık = [17.5, 17.8, 13.6, 7.3, 0.4, -6.9]

aylar = ["Temmuz","Ağustos","Eylül","Ekim","Kasım","Aralık"]

plt.plot(aylar,edirne_sıcaklık,"g",marker=".", markerfacecolor="black")
plt.plot(aylar,sinop_sıcaklık, "r",marker=".", markerfacecolor="black")
plt.plot(aylar,antalya_sıcaklık, "b",marker=".", markerfacecolor="black")
plt.plot(aylar,kars_sıcaklık, "y",marker=".", markerfacecolor="black")
plt.ylabel("Sıcaklık (°C)")
plt.xlabel("Aylar")
plt.title("İLLERİN SICAKLIK (°C) GRAFİĞİ")
plt.legend(["Edirne °C","Sinop °C","Antalya °C","Kars °C"])
plt.ylim(-10,30)
plt.xlim(0,5)
plt.grid(True)
plt.show()

```
